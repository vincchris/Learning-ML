```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print("KEYS DATASETS:")
print(iris.keys())

print("\nFEATURE NAME:")
print(iris.feature_names)

print("\nTARGET NAME:")
print(iris.target_names)
```

```python
X = iris.data
y = iris.target

print("FEATURE:", X.shape)
print("LABEL:", y.shape)
```

```python
print("\nFIRST DATA FEATURE:")
print(X[:5])

print("\nFIRST LABEL:")
print(y[:5])
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y

print("\nDATAFRAME HEAD:")
print(df.head())
```
